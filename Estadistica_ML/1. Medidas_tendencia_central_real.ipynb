# Medidas de Tendencia Central Aplicadas a Machine Learning

En este notebook se abordan las medidas de tendencia central: **media**, **mediana** y **moda**. Estas medidas son fundamentales para entender la distribución de los datos y, en el contexto de Machine Learning, permiten identificar el comportamiento central de las características y ayudar en el preprocesamiento de los datos.

A continuación se muestran las definiciones y fórmulas de cada una de estas medidas, seguidas de un ejemplo real aplicado al dataset **Iris**.

## 1. Conceptos Teóricos

### a) Media (Promedio)

La **media** es el valor promedio de un conjunto de datos y se calcula como:

$$ \bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_i $$

donde:

- **\( n \)** es el número de observaciones.
- **\( $x_i$ \)** es el valor de la i-ésima observación.

### b) Mediana

La **mediana** es el valor central de un conjunto de datos ordenado. Si el número de observaciones es impar, es el valor medio; si es par, es el promedio de los dos valores centrales.

### c) Moda

La **moda** es el valor que ocurre con mayor frecuencia en el conjunto de datos. Es especialmente útil en variables categóricas, pero puede aplicarse también a variables numéricas discretas.

## Ejemplos de Cálculo de la Media

A continuación se presentan **tres ejemplos** de cálculo de la media utilizando Python:

```python
import numpy as np

# Ejemplo 1: Media de un arreglo simple
datos1 = np.array([10, 20, 30, 40, 50])
media1 = np.mean(datos1)
print('Ejemplo 1 - Media:', media1)

# Ejemplo 2: Media de edades
edades = np.array([22, 25, 25, 30, 34, 40, 40, 40, 45])
media_edades = np.mean(edades)
print('Ejemplo 2 - Media de edades:', media_edades)

# Ejemplo 3: Media de puntuaciones en un examen
puntuaciones = np.array([85, 90, 78, 92, 88, 76, 95])
media_puntuaciones = np.mean(puntuaciones)
print('Ejemplo 3 - Media de puntuaciones:', media_puntuaciones)
```

## Ejemplos de Cálculo de la Mediana

A continuación se presentan **tres ejemplos** para el cálculo de la mediana:

```python
import numpy as np

# Ejemplo 1: Mediana de un arreglo simple
datos1 = np.array([10, 20, 30, 40, 50])
mediana1 = np.median(datos1)
print('Ejemplo 1 - Mediana:', mediana1)

# Ejemplo 2: Mediana de edades
edades = np.array([22, 25, 25, 30, 34, 40, 40, 40, 45])
mediana_edades = np.median(edades)
print('Ejemplo 2 - Mediana de edades:', mediana_edades)

# Ejemplo 3: Mediana de puntuaciones en un examen
puntuaciones = np.array([85, 90, 78, 92, 88, 76, 95])
mediana_puntuaciones = np.median(puntuaciones)
print('Ejemplo 3 - Mediana de puntuaciones:', mediana_puntuaciones)
```

## Ejemplos de Cálculo de la Moda

A continuación se presentan **tres ejemplos** para el cálculo de la moda utilizando `scipy.stats.mode`:

Recuerda que la moda es el valor que más se repite en el conjunto de datos.

```python
import numpy as np
from scipy import stats

# Ejemplo 1: Moda de un arreglo simple
datos1 = np.array([10, 20, 20, 30, 40, 50])
moda1 = stats.mode(datos1, keepdims=True).mode[0]
print('Ejemplo 1 - Moda:', moda1)

# Ejemplo 2: Moda en edades
edades = np.array([22, 25, 25, 30, 34, 40, 40, 40, 45])
moda_edades = stats.mode(edades, keepdims=True).mode[0]
print('Ejemplo 2 - Moda de edades:', moda_edades)

# Ejemplo 3: Moda de calificaciones
calificaciones = np.array([70, 80, 80, 85, 90, 80, 95])
moda_calificaciones = stats.mode(calificaciones, keepdims=True).mode[0]
print('Ejemplo 3 - Moda de calificaciones:', moda_calificaciones)
```

## Técnicas de Sustitución para Datos Faltantes

En Machine Learning es común encontrar conjuntos de datos con valores faltantes. Para manejarlos, existen varias técnicas de imputación (sustitución) formales, entre las cuales se destacan:

- **Imputación por la Media:** Se reemplaza el valor faltante por la media de la columna.
  - Fórmula: $$ x_{imputado} = \bar{x} $$

- **Imputación por la Mediana:** Se reemplaza el valor faltante por la mediana de la columna.
  - Fórmula: $$ x_{imputado} = \tilde{x} $$

- **Imputación por la Moda:** Se reemplaza el valor faltante por la moda de la columna.
  - Fórmula: $$ x_{imputado} = moda(x) $$

Estas técnicas permiten mantener la integridad de los datos y evitar la pérdida de información. En Python, librerías como `pandas` y `scikit-learn` ofrecen funciones para realizar estas imputaciones de manera sencilla.

## Ejemplo de Imputación con Pandas

En el siguiente ejemplo se muestra cómo imputar valores faltantes utilizando la media, mediana y moda en un DataFrame de `pandas`.

```python
import pandas as pd
import numpy as np

# Crear un DataFrame de ejemplo con valores faltantes
data = {
    'Edad': [25, 30, np.nan, 22, 28, np.nan, 35],
    'Puntuacion': [85, np.nan, 78, 92, np.nan, 76, 95]
}
df = pd.DataFrame(data)
print('DataFrame original:')
print(df)

# Imputación utilizando la media
df_imputacion_media = df.fillna(df.mean(numeric_only=True))
                        # numeric_only - determina si la media se aplica solo a columnas numéricas o no
print('\nDataFrame con imputación por la media:')
print(df_imputacion_media)

# Imputación utilizando la mediana
df_imputacion_mediana = df.fillna(df.median(numeric_only=True))
print('\nDataFrame con imputación por la mediana:')
print(df_imputacion_mediana)

# Para la moda, se utiliza el método de imputación individual
df_imputacion_moda = df.copy()  # se hace una copia del dataframe para no modificar los valores originales
for columna in df_imputacion_moda.columns:
    # Calcular la moda, si existen varios valores, se toma el primero
    moda = df_imputacion_moda[columna].mode()
    if not moda.empty:
        df_imputacion_moda[columna].fillna(moda[0], inplace=True)
                                    # fillna - reemplaza los nan del df

print('\nDataFrame con imputación por la moda:')
print(df_imputacion_moda)
```

## 2. Ejemplo Real: Análisis del Dataset Iris

En este ejemplo se utilizará el dataset **Iris** para calcular las medidas de tendencia central de la característica **'sepal length (cm)'**. Este análisis es útil para comprender la distribución de esta característica, lo cual es fundamental en el preprocesamiento y la selección de características en proyectos de Machine Learning.

### Pasos a seguir:

1. Cargar el dataset Iris.
2. Calcular la media, mediana y moda de la característica **'sepal length (cm)'**.
3. Visualizar la distribución de la característica.

¡Comencemos!

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
from scipy import stats

# Cargar el dataset Iris
iris = datasets.load_iris()
df_iris = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Mostramos las primeras filas
print('Primeras 5 filas del dataset Iris:')
display(df_iris.head())

# Seleccionamos la característica 'sepal length (cm)'
sepal_length = df_iris['sepal length (cm)']

# Cálculo de la Media
media = np.mean(sepal_length)
print('\nMedia (Promedio) de Sepal Length:', round(media, 3))

# Cálculo de la Mediana
mediana = np.median(sepal_length)
print('Mediana de Sepal Length:', round(mediana, 3))

# Cálculo de la Moda
moda = stats.mode(sepal_length, keepdims=True).mode[0]
print('Moda de Sepal Length:', round(moda, 3))

# Visualización: Histograma con KDE
plt.figure(figsize=(12, 6))
sns.histplot(sepal_length, kde=True, bins=20, color='lightblue')
                    # kde - superpone una curva suavizada que representa la densidad de probabilidad de los datos
                        # IMPORTANTE solo aplica en histogramas
plt.title('Distribución de Sepal Length (cm)')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frecuencia')
plt.axvline(media, color='orange', linestyle='dashed', linewidth=1.5, label=f'Media: {media:.2f}')
plt.axvline(mediana, color='darkblue', linestyle='dashed', linewidth=1.5, label=f'Mediana: {mediana:.2f}')
                                    # .2f formatea números decimales con dos cifras después del punto decimal
plt.legend()
plt.show()
```

```python
# Realizamos lo mismo pero con la longitud de los pétalos
# Seleccionamos la característica 'pepal length (cm)'
petal_length = df_iris['petal length (cm)']

# Cálculo de la Media
petal_media = np.mean(petal_length)
print('\nMedia (Promedio) de Petal Length:', round(petal_media, 3))

# Cálculo de la Mediana
petal_mediana = np.median(petal_length)
print('Mediana de Petal Length:', round(petal_mediana, 3))

# Cálculo de la Moda
petal_moda = stats.mode(petal_length, keepdims=True).mode[0]
print('Moda de Petal Length:', round(petal_moda, 3))

# Visualización: Histograma con KDE
plt.figure(figsize=(12, 6))
sns.histplot(petal_length, kde=True, bins=20, color='plum')
plt.title('Distribución de Petal Length (cm)')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frecuencia')
plt.axvline(petal_media, color='violet', linestyle='dashed', linewidth=1.5, label=f'Media: {petal_media:.2f}')
plt.axvline(petal_mediana, color='purple', linestyle='dashed', linewidth=1.5, label=f'Mediana: {petal_mediana:.2f}')
plt.legend()
plt.show()
```

```python
# Seleccionamos la característica 'sepal width (cm)'
sepal_width = df_iris['sepal width (cm)']

# Cálculo de la Media
media_sw = np.mean(sepal_width)
print('\nMedia (Promedio) de Sepal width:', round(media_sw, 3))

# Cálculo de la Mediana
mediana_sw = np.median(sepal_width)
print('Mediana de Sepal width:', round(mediana_sw, 3))

# Cálculo de la Moda
moda_sw = stats.mode(sepal_width, keepdims=True).mode[0]
print('Moda de Sepal width:', round(moda_sw, 3))

# Visualización: Histograma con KDE
plt.figure(figsize=(12, 6))
sns.histplot(sepal_width, kde=True, bins=20, color='lightblue')
plt.title('Distribución de Sepal width (cm)')
plt.xlabel('Sepal width (cm)')
plt.ylabel('Frecuencia')
plt.axvline(media_sw, color='orange', linestyle='dashed', linewidth=1.5, label=f'Media: {media_sw:.2f}')
plt.axvline(mediana_sw, color='darkblue', linestyle='dashed', linewidth=1.5, label=f'Mediana: {mediana_sw:.2f}')
plt.legend()
plt.show()
```

```python
# Seleccionamos la característica 'petal width (cm)'
petal_width = df_iris['petal width (cm)']

# Cálculo de la Media
media_pw = np.mean(petal_width)
print('\nMedia (Promedio) de Petal width:', round(media_pw, 3))

# Cálculo de la Mediana
mediana_pw = np.median(petal_width )
print('Mediana de Petal width:', round(mediana_pw, 3))

# Cálculo de la Moda
moda_pw = stats.mode(petal_width, keepdims=True).mode[0]
print('Moda de Petal width:', round(moda_pw, 3))

# Visualización: Histograma con KDE
plt.figure(figsize=(12, 6))
sns.histplot(petal_width, kde=True, bins=20, color='plum')
plt.title('Distribución de Petal width (cm)')
plt.xlabel('Petal width (cm)')
plt.ylabel('Frecuencia')
plt.axvline(media_pw, color='violet', linestyle='dashed', linewidth=1.5, label=f'Media: {media_pw:.2f}')
plt.axvline(mediana_pw, color='purple', linestyle='dashed', linewidth=1.5, label=f'Mediana: {mediana_pw:.2f}')
plt.legend()
plt.show()
```

Estas medidas de tendencia central permiten obtener una idea clara de la posición central de los datos:

La comprensión de estas medidas es fundamental en el preprocesamiento de datos para Machine Learning, ya que ayudan a identificar y corregir problemas de sesgo, a realizar imputaciones de valores faltantes y a escalar los datos de manera adecuada.
