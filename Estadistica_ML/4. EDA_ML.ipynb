# Análisis Exploratorio de Datos (EDA) Aplicado a Machine Learning

El Análisis Exploratorio de Datos (EDA) es el proceso de examinar y visualizar un conjunto de datos para entender sus características principales. Es fundamental en Machine Learning, ya que permite detectar patrones, valores atípicos, correlaciones y la distribución de las variables, lo cual es clave para el preprocesamiento y la construcción de modelos predictivos.

En este Notebook se presentarán:

- **Conceptos teóricos** con fórmulas en LaTeX y descripción de variables.
- **Ejemplos prácticos**: histograma, boxplot y scatter plot, con su análisis interpretativo.
- Un **ejemplo real** utilizando el dataset *Iris* para ilustrar el EDA aplicado a Machine Learning.

## 1. Conceptos Teóricos del EDA

### a) Medidas de Tendencia Central

Estas medidas resumen el comportamiento central de los datos.

**Media (\($\bar{x}$\))**: valor promedio de un conjunto de datos.

$$\bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_i$$

donde:
- \(n\): número de observaciones.
- \($x_i$\): valor de la i-ésima observación.

**Mediana**: valor central cuando los datos están ordenados.

**Moda**: valor o valores que ocurren con mayor frecuencia.

### b) Medidas de Dispersión

**Varianza (\($\sigma^2$\))**: mide la dispersión de los datos respecto a la media.

$$\sigma^2 = \frac{1}{n} \sum_{i=1}^{n} (x_i-\bar{x})^2$$

**Desviación Estándar (\($\sigma$\))**: es la raíz cuadrada de la varianza, y se expresa en las mismas unidades que los datos.

$$\sigma = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (x_i-\bar{x})^2}$$

### c) Boxplot

El boxplot (diagrama de caja) resume la distribución de los datos a través de cinco números:

- **Mínimo** (sin contar outliers).
- **Primer cuartil (Q1)**: el 25% inferior de los datos.
- **Mediana (Q2)**: el valor central.
- **Tercer cuartil (Q3)**: el 75% inferior de los datos.
- **Máximo** (sin contar outliers).

El **IQR (Interquartile Range)** se define como:

$$ IQR = Q3 - Q1 $$

y se utiliza para identificar valores atípicos (outliers).

### d) Scatter Plot

El scatter plot (gráfico de dispersión) muestra la relación entre dos variables, permitiendo detectar correlaciones y agrupamientos.

## 2. Ejemplos Prácticos de EDA

A continuación se presentan ejemplos prácticos para visualizar y analizar los datos mediante diferentes gráficas.

### Ejemplo 1: Histograma

El histograma es una gráfica que muestra la distribución de una variable numérica. Es útil para identificar la forma de la distribución (por ejemplo, simetría, sesgo) y detectar posibles valores atípicos.

En el siguiente ejemplo se genera un histograma de datos aleatorios que siguen una distribución normal.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generar datos con distribución normal
mu, sigma = 0, 1
datos = np.random.normal(mu, sigma, 1000)

# Graficar el histograma con KDE
plt.figure(figsize=(8, 4))
sns.histplot(datos, bins=30, kde=True, color='plum')
                            # kde - superpone una curva de densidad suave sobre el histograma para 
                            # representar la distribución de los datos de una manera más continua.
plt.title('Histograma de Datos (Distribución Normal)')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()

# Análisis:
# La gráfica muestra una distribución simétrica centrada en 0, con la curva KDE ayudando a visualizar
# la forma de la distribución N(0,1).
```

### Ejemplo 2: Boxplot

El boxplot resume la distribución de los datos mostrando la mediana, los cuartiles y los posibles valores atípicos.

En el siguiente ejemplo se crea un boxplot para un conjunto de datos generados a partir de una distribución normal.

```python
# Generar datos aleatorios
datos_box = np.random.normal(50, 10, 200)

# Graficar el boxplot
plt.figure(figsize=(8, 4))
sns.boxplot(x=datos_box, color='lightblue')
plt.title('Boxplot de Datos Aleatorios')
plt.xlabel('Valor')
plt.show()

# Análisis:
# El boxplot muestra la mediana (línea central) y los cuartiles Q1 y Q3. Los 'bigotes' se extienden 
# hasta 1.5 veces el IQR, y los puntos fuera de este rango se consideran outliers.
```

### Ejemplo 3: Scatter Plot

El scatter plot es útil para analizar la relación entre dos variables. Permite identificar tendencias, correlaciones y agrupamientos.

En el siguiente ejemplo se genera un scatter plot de dos variables aleatorias.

```python
# Generar datos para dos variables
x_scatter = np.random.rand(100) * 100
                # rand genera números con distribución uniforme en el intervalo [0, 1]
y_scatter = 0.5 * x_scatter + np.random.randn(100) * 10
                                    # randn genera números con distribución normal estándar, ie, N(0, 1)

# Graficar el scatter plot
plt.figure(figsize=(8, 4))
plt.scatter(x_scatter, y_scatter, color='coral')
plt.title('Scatter Plot: Relación entre dos Variables')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.show()

# Análisis:
# Se observa una tendencia lineal positiva entre X e Y, aunque existe dispersión debido al ruido en los datos.
```

## 3. Ejemplo Real Aplicado a Machine Learning: Análisis del Dataset Iris

En este ejemplo real se aplicará EDA al dataset **Iris** para comprender las distribuciones y relaciones entre sus características antes de utilizarlo en un modelo de Machine Learning.

### Pasos del análisis:

1. **Carga y resumen de datos:** Se visualizan las primeras filas y se obtienen estadísticas descriptivas.
2. **Histograma:** Se analiza la distribución de la característica *sepal length (cm)*.
3. **Boxplot:** Se examina la distribución y se identifican outliers en *petal width (cm)*.
4. **Scatter Plot:** Se exploran las relaciones entre *sepal length (cm)* y *petal length (cm)*.

¡Comencemos!

```python
import pandas as pd
from sklearn import datasets

# Cargar el dataset Iris
iris = datasets.load_iris()
df_iris = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df_iris['target'] = iris.target

# Visualizar las primeras filas y estadísticas descriptivas
print('Primeras 5 filas del dataset Iris:')
print(df_iris.head())

print('\nEstadísticas descriptivas:')
print(df_iris.describe())

# Histograma de 'sepal length (cm)'
plt.figure(figsize=(8, 4))
sns.histplot(df_iris['sepal length (cm)'], bins=20, kde=True, color='skyblue')
plt.title('Distribución de Sepal Length (cm)')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frecuencia')
plt.show()

# Análisis del Histograma:
# La distribución muestra que la mayoría de las muestras se agrupan en un rango central, lo que indica una distribución moderadamente simétrica.

# Boxplot de 'petal width (cm)'
plt.figure(figsize=(8, 4))
sns.boxplot(x=df_iris['petal width (cm)'], color='lightgreen')
plt.title('Boxplot de Petal Width (cm)')
plt.xlabel('Petal Width (cm)')
plt.show()

# Análisis del Boxplot:
# Se observa la mediana (línea central), Q1 y Q3. Los bigotes indican el rango de los datos sin outliers; 
# si existen puntos fuera de estos, se consideran valores atípicos.

# Scatter Plot: Relación entre 'sepal length (cm)' y 'petal length (cm)'
plt.figure(figsize=(8, 4))
plt.scatter(df_iris['sepal length (cm)'], df_iris['petal length (cm)'], c=df_iris['target'], cmap='viridis')
plt.title('Relación entre Sepal Length y Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.colorbar(label='Target')
plt.show()

# Análisis del Scatter Plot:
# Se aprecia una correlación positiva entre 'sepal length (cm)' y 'petal length (cm)', 
# lo que es indicativo de que estas características podrían ser útiles para la clasificación.
```

```python
# Histograma de 'sepal width (cm)'
plt.figure(figsize=(8, 4))
sns.histplot(df_iris['sepal width (cm)'], bins=20, kde=True, color='plum')
plt.title('Distribución de Sepal Width (cm)')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frecuencia')
plt.show()

# Análisis del Histograma:
# La distribución muestra que la mayoría de las muestras se agrupan en un rango central, 
# lo que indica una distribución moderadamente simétrica y centrada en la media.

# Boxplot de 'sepal width (cm)'
plt.figure(figsize=(8, 4))
sns.boxplot(x=df_iris['sepal width (cm)'], color='pink')
plt.title('Boxplot de Sepal Width (cm)')
plt.xlabel('Sepal Width (cm)')
plt.show()

# Análisis del Boxplot:
# Se observa la mediana (línea central), Q1 y Q3. Los bigotes indican el rango de los datos sin outliers;
# si existen puntos fuera de estos, se consideran valores atípicos.

# Scatter Plot: Relación entre 'sepal width (cm)' y 'petal width (cm)'
plt.figure(figsize=(8, 4))
plt.scatter(df_iris['sepal width (cm)'], df_iris['petal width (cm)'], c=df_iris['target'], cmap='viridis')
plt.title('Relación entre Sepal width y Petal width')
plt.xlabel('Sepal width (cm)')
plt.ylabel('Petal width (cm)')
plt.colorbar(label='Target')
plt.show()

# Análisis del Scatter Plot:
# Se parecia que la relación entre 'sepal width (cm)' y 'petal width (cm)', ya no es tan clara
# lo que es indicativo de que estas características podrían no ser útiles para la clasificación.
```
