# Matplolib

```python
# importing the required module
import matplotlib.pyplot as plt
```

```python
# create data for plotting
x_values = [0, 1, 2, 3, 4, 5]
y_values = [0, 1, 4, 9, 16, 25]

# the default graph style for plot is a line
plt.plot(x_values,y_values)

# display the graph
plt.show()
```

```python
plt.plot([1, 2, 3, 4])
plt.ylabel('some numbers')
plt.show()
```

```python
y1 = []
y2 = []
x = range(-100, 100, 10) # Secuencia de números entre -100 y 100, con separación de 10
for i in x: y1.append(i**2)
for i in x: y2.append(-i**2)

plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel("x")  # Colocar etiqueta al eje x
plt.ylabel("y")  # Colocar etiqueta al eje y
plt.ylim(-2000, 2000)
plt.axhline(0)  # horizontal line centrada en 0
plt.axvline(0)  # vertical line centrada en 0

plt.savefig("quad.png") # Guarda la imagen como quad en la carpeta
plt.show()
```

```python
# x axis values
x = [1, 2, 3]
# corresponding y axis values
y = [2, 4, 1]

# plotting the points
plt.plot(x, y)

# NAMING AXIS
# naming the x axis
plt.xlabel("x - axis")
# naming the y axis
plt.ylabel("y - axis")

# giving a title to my graph
plt.title("mi first graph!")

# function to show the plot
plt.show()
```

```python
# line 1 points
x1 = [1, 2, 3]
y1 = [2, 4, 1]
# plotting the line 1 points
plt.plot(x1, y1, label = "line 1")

# line 2 points
x2 = [1, 2, 3]
y2 = [4, 1, 3]
# plotting the line 2 points
plt.plot(x2, y2, label = "line 2")

# naming the x axis
plt.xlabel("x - axis")
# naming the y axis
plt.ylabel("y - axis")
# giving a title to my graph
plt.title("Two lines on same graph")

# show legend on the plot
plt.legend()

# function to show the plot
plt.show()
```
