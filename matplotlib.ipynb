# Matplolib

```python
import matplotlib.pyplot as plt
```

```python
# create data for plotting
x_values = [0, 1, 2, 3, 4, 5]
y_values = [0, 1, 4, 9, 16, 25]

# the default graph style for plot is a line
plt.plot(x_values,y_values)

# display the graph
plt.show()
```

```python
plt.plot([1, 2, 3, 4])
plt.ylabel('some numbers')
plt.show()
```

```python
y1 = []
y2 = []
x = range(-100, 100, 10) # Secuencia de números entre -100 y 100, con separación de 10
for i in x: y1.append(i**2)
for i in x: y2.append(-i**2)

plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel("x")  # Colocar etiqueta al eje x
plt.ylabel("y")  # Colocar etiqueta al eje y
plt.ylim(-2000, 2000)
plt.axhline(0)  # horizontal line centrada en 0
plt.axvline(0)  # vertical line centrada en 0

plt.savefig("quad.png") # Guarda la imagen como quad en la carpeta
plt.show()
```
